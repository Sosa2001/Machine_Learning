```python
import numpy as np
import matplotlib.pyplot as plt
```

# Example with $h_\theta(x) = \theta\cdot x = \theta_1 x_1$

```python
X = np.linspace(0, 5, 6)
y = X
h = y

print(X)
print(y)
```

```python
plt.figure(figsize=(5,5))
plt.xlabel('X', fontsize = 12)
plt.ylabel('y', fontsize = 12)
plt.plot(X, h, 'g--')
plt.plot(X, y, 'ro', markersize = 5)
plt.grid(True)
plt.legend(['h', 'y'])
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.xlabel('X', fontsize = 12)
plt.ylabel('y', fontsize = 12)

theta1 = 0.5
h = theta1 * X
plt.plot(X, h, 'g--')

plt.plot(X, y, 'ro', markersize = 5)
plt.grid(True)
plt.legend(['h', 'y'])
plt.show()
```

```python
m = X.size
theta1 = np.linspace(-3, 5, 100)

h = [theta*X for theta in theta1]
J = (1/(2*m)*((h-y)**2).sum(axis = 1)) #MSE

plt.xlabel(r'$\theta_1$', fontsize = 12)
plt.ylabel(r'$J\theta_1$', fontsize = 12)
plt.plot(theta1, J, 'b-', linewidth = 2)
plt.grid(True)
plt.show()
```

```python
m = X.size
theta1 = np.linspace(-3, 5, 100)

h = [theta*X for theta in theta1]
J = np.sqrt((1/(2*m)*((h-y)**2).sum(axis = 1))) #MSE

plt.xlabel(r'$\theta_1$', fontsize = 12)
plt.ylabel(r'$J\theta_1$', fontsize = 12)
plt.plot(theta1, J, 'b-', linewidth = 2)
plt.grid(True)
plt.show()
```

```python
m = X.size
theta1 = np.linspace(-3, 5, 100)

h = [theta*X for theta in theta1]
J = np.sqrt((1/(2*m)*(np.abs(h-y)**2).sum(axis = 1))) #MSE

plt.xlabel(r'$\theta_1$', fontsize = 12)
plt.ylabel(r'$J\theta_1$', fontsize = 12)
plt.plot(theta1, J, 'b-', linewidth = 2)
plt.grid(True)
plt.show()
```

# Example with $h_\theta(x) = \theta\cdot x = \theta_0 + \theta_1 x_1$

```python
X = np.linspace(-2, 2, 10)
y = 2.5 +5*X
```

```python
m = X.size

theta0 = np.linspace(-12.5, 17.5, 100)
theta1 = np.linspace(-10, 20, 100)

h = np.array([[th0 +th1*X for th0 in theta0] for th1 in theta1])
J = (1/(2*m)*((h-y)**2).sum(axis = 2)) #MSE

fig = plt.figure()
ax = plt.axes(projection='3d')
th0, th1 = np.meshgrid(theta0, theta1)
surf = ax.plot_surface(th0, th1, J, cmap='coolwarm')

ax.set_xlabel(r'$\theta_0$', fontsize=12)
ax.set_ylabel(r'$\theta_1$', fontsize=12)
ax.set_zlabel(r'$J(\theta_0, \theta_1)$', fontsize=12)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.view_init(20, 45)
plt.show()
```

```python
plt.contour(th0, th1, J, 50, cmap='coolwarm')
plt.xlabel(r'$\theta_0$', fontsize=12)
plt.ylabel(r'$\theta_1$', fontsize=12)
plt.colorbar()
plt.show()
```
